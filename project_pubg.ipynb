#### 22. Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot it on a histogram and comment on its distribution.

```python
import seaborn as sns
mean_dd = []
for i in range(100):
    mean_damage = df['damageDealt'].sample(50).mean()
    mean_dd.append(mean_damage)
mean_dd = pd.DataFrame(mean_dd)
print(mean_dd)
mean_dd.hist()
plt.xticks(rotation=70);
```

After taking a sample of size 50 from the column damageDealt for 100 times we can clearly see that the mean in most of the cases is in the range of 100 - 160
